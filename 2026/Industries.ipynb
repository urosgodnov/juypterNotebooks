## Data Preparation and Cleaning

Build mapping A*10 industry ↔ NACE first letter

In nama_10_a10 the A*10 industries are in nace_r2:

A, B-E, C, F, G-I, J, K, L, M-N, O-Q, R-U

In SBS you have detailed NACE (C10, C11, …).
We map the first letter of SBS NACE to the A*10 group.

```r
start_date <- "2005-01-01"

# Prepare GDP Data
gdp_clean <- gdp_raw %>%
  filter(geo %in% eu_countries,    # Filter for ALL countries in your list
         unit == "CLV10_MEUR",
         na_item == "B1GQ") %>%
  rename(time=TIME_PERIOD) %>%
  select(geo, time, values) %>%
  rename(GDP = values) %>%
  mutate(year = year(time)) %>%
  filter(time >= as.Date(start_date))%>%
    select(-time)

# Prepare Value Added
va_clean <- va_raw %>%
  filter(geo %in% eu_countries,
         unit == "CLV10_MEUR",
         na_item == "B1G",
         nace_r2 != "TOTAL") %>%
   rename(time=TIME_PERIOD) %>%
  select(geo, time, nace_r2, values) %>%
  rename(ValueAdded = values) %>%
  filter(time >= as.Date(start_date))

# Prepare Employment
emp_clean <- emp_raw %>%
  filter(geo %in% eu_countries,
         unit == "THS_PER",
         na_item == "EMP_DC",
         nace_r2 != "TOTAL") %>%
   rename(time=TIME_PERIOD) %>%
  select(geo, time, nace_r2, values) %>%
  rename(Employment = values) %>%
  filter(time >= as.Date(start_date))

```

To analyze the connection between companies and the aggregate (GDP), you should select these three key variables:

* For Sales (Prodaja): NETTUR_MEUR

Meaning: Net Turnover in Million Euros. This represents the total sales of the companies.

* For Value Added (Dodana vrednost): AV_MEUR

Meaning: Value Added in Million Euros. This represents the wealth created by the companies (comparable to GDP).

* For Employment (Zaposlenost): EMP_NR

Meaning: Number of persons employed. This is the headcount of people working in these sectors.

```r
sbs_clean <- sbs_new_raw %>%
  mutate(
    indic_sbs = case_when(
      indic_sb == "V12110" ~ "NETTUR_MEUR",
      indic_sb == "V12150" ~ "AV_MEUR",
      indic_sb == "V16110" ~ "EMP_NR",
      TRUE ~ NA_character_
    )
  ) %>%
  filter(!is.na(indic_sbs)) %>%        # samo 3 kazalniki
  rename(time = TIME_PERIOD) %>%
  select(-indic_sb) %>%
  # 2) IZPELJEMO DIVIZIJO IZ PODROBNE KODE (H4910 -> H49, J6203 -> J62, ...)
  mutate(
    nace_division = str_sub(nace_r2, 1, 3)
  ) %>%
  # obdržimo samo prave divizije (črka + 2 števki)
  filter(str_detect(nace_division, "^[A-Z][0-9]{2}$"))

# 3) Agregacija na ravni divizij
sbs_aggregated <- sbs_clean %>%
  group_by(geo, time, nace_division, indic_sbs) %>%
  summarise(values = sum(values, na.rm = TRUE), .groups = "drop")

# 4) Slovar NACE (uradne labele) – filtriramo samo divizije
nace_dict <- get_eurostat_dic("nace_r2") %>%
  select(code_name, full_name) %>%
  rename(nace_division = code_name, label_text = full_name) %>%
  filter(str_detect(nace_division, "^[A-Z][0-9]{2}$"))

# 5) Wide format + labela
sbs_wide <- sbs_aggregated %>%
  pivot_wider(names_from = indic_sbs, values_from = values) %>%
  left_join(nace_dict, by = "nace_division") %>%
  mutate(
    label = paste(nace_division, label_text, sep = " - "),
    year = year(time)
  )%>%
    select(-time)
```

Calculate Indices (Base Year = 100)


```r
rebase_to_100 <- function(x) {
  if(all(is.na(x))) return(x)
  # Use first non-NA value as base, or just first value
  first_val <- x[which(!is.na(x))[1]]
  if(is.na(first_val) || first_val == 0) return(x)
  x / first_val * 100
}
```

```r
# CRITICAL: We must group by 'geo' and 'nace_r2' so indices are calculated per country/sector
df_final <- sbs_wide %>%
  inner_join(gdp_clean, by = c("geo", "year")) %>%
  group_by(geo, nace_division) %>%
  arrange(year, .by_group = TRUE) %>%
  mutate(
    Index_GDP         = rebase_to_100(GDP),
    Index_Sales       = rebase_to_100(NETTUR_MEUR),
    Index_ValueAdded  = rebase_to_100(AV_MEUR),
    Index_Employment  = rebase_to_100(EMP_NR)
  ) %>%
  ungroup() %>%
  mutate(
    # all rows are now divisions (H49, I55, J62, ...)
    data_level  = "Division",
    level1_code = substr(nace_division, 1, 1),
    macro_sector = case_when(
      level1_code == "A" ~ "Agriculture",
      level1_code %in% c("B", "C", "D", "E") ~ "Industry",
      level1_code == "F" ~ "Construction",
      level1_code %in% c("G", "H", "I") ~ "Trade, Transport, Accom.",
      level1_code == "J" ~ "Info & Comm",
      level1_code == "K" ~ "Finance",
      level1_code == "L" ~ "Real Estate",
      level1_code %in% c("M", "N") ~ "Professional/Sci",
      level1_code %in% c("O", "P", "Q") ~ "Public Admin/Health",
      level1_code %in% c("R", "S", "T", "U") ~ "Arts/Other",
      TRUE ~ "Other"
    )
  )

head(df_final)
```

## Analysis


#### SENSITIVITY (CORRELATION)

```r
# ------------------------------------------------------------------------------
# FIX: Define a "Safe" Correlation Function
# This prevents the crash when a sector (like Finance) has no Sales data.
# ------------------------------------------------------------------------------
safe_cor <- function(x, y) {
  # Count how many years have BOTH x and y data
  valid_points <- sum(!is.na(x) & !is.na(y))

  # We need at least 3 years of data to calculate a correlation
  if (valid_points < 3) {
    return(NA_real_) # Return "NA" (Blank) instead of crashing
  } else {
    return(cor(x, y, use = "complete.obs"))
  }
}

target_country <- "SI"

table_sensitivity <- df_final %>%
  filter(geo == target_country) %>%   # no data_level filter for now
  group_by(macro_sector, label) %>%
  summarise(
    Corr_Sales_GDP   = round(safe_cor(Index_Sales, Index_GDP), 2),
    Avg_Sales_MEUR   = round(mean(NETTUR_MEUR, na.rm = TRUE), 0),
    Corr_Emp_GDP     = round(safe_cor(Index_Employment, Index_GDP), 2),
    .groups = "drop"
  ) %>%
  filter(!is.na(Corr_Sales_GDP)) %>%
  arrange(desc(Corr_Sales_GDP)) %>%
  select(label, macro_sector, Corr_Sales_GDP, Corr_Emp_GDP, Avg_Sales_MEUR)

# PRINT TABLE
print(paste("--- TABLE 1: GDP SENSITIVITY (Top 20 Most Correlated Sectors in", target_country, ") ---"))
print("(1.00 = Perfectly Cyclical / Moves with GDP.  Near 0 = Independent)")

table_sensitivity %>%
  head()


table_sensitivity %>%
  tail()
```

#### GROWTH & MULTIPLIERS

```r
table_growth <- df_final %>%
  # if you renamed/changed data_level, drop that filter or adapt it
  filter(geo == target_country) %>%
  arrange(label, macro_sector, year) %>%   # ensure time order inside groups
  group_by(label, macro_sector) %>%
  summarise(
    # deltas vs base index = 100
    Sales_Delta = last(Index_Sales) - 100,
    Emp_Delta   = last(Index_Employment) - 100,
    GDP_Delta   = last(Index_GDP) - 100,

    Sales_Growth_Pct = round(Sales_Delta, 1),
    Emp_Growth_Pct   = round(Emp_Delta, 1),
    GDP_Growth_Pct   = round(GDP_Delta, 1),

    Sales_Multiplier = ifelse(is.na(GDP_Delta) | GDP_Delta == 0,
                              NA_real_,
                              round(Sales_Delta / GDP_Delta, 2)),
    Emp_Multiplier   = ifelse(is.na(GDP_Delta) | GDP_Delta == 0,
                              NA_real_,
                              round(Emp_Delta / GDP_Delta, 2)),
    .groups = "drop"
  ) %>%
  filter(!is.na(Sales_Growth_Pct)) %>%
  arrange(desc(Sales_Multiplier)) %>%
  select(label, macro_sector,
         Sales_Growth_Pct, Sales_Multiplier,
         Emp_Growth_Pct,   Emp_Multiplier)

print(paste(
  "--- TABLE 2: GROWTH & MULTIPLIERS (Sales vs Employment in",
  target_country, ") ---"
))
print("(Sales_Multiplier > 1: Revenue beat GDP. Emp_Multiplier > 1: Hiring beat GDP.)")

table_growth %>%
  head() %>%
  gt() %>%
  gts()
```

#### LABOR PRODUCTIVITY (Efficiency)

```r
table_productivity <- df_final %>%
  filter(geo == target_country) %>%             # drop data_level filter
  # ensure chronological order inside each sector
  arrange(label, macro_sector, year) %>%        # use `time` if you don't have `year`
  # drop bad rows before computing productivity
  filter(!is.na(AV_MEUR), !is.na(EMP_NR), EMP_NR > 0) %>%
  mutate(
    # AV_MEUR is in million EUR → convert to EUR per worker
    Prod_EUR = (AV_MEUR * 1e6) / EMP_NR
  ) %>%
  group_by(label, macro_sector) %>%
  summarise(
    Prod_Start_EUR      = round(first(Prod_EUR), 0),
    Prod_End_EUR        = round(last(Prod_EUR), 0),
    Efficiency_Gain_Pct = round((last(Prod_EUR) / first(Prod_EUR) - 1) * 100, 1),
    .groups = "drop"
  ) %>%
  filter(!is.na(Efficiency_Gain_Pct)) %>%
  arrange(desc(Efficiency_Gain_Pct)) %>%
  select(label, macro_sector, Prod_End_EUR, Efficiency_Gain_Pct)

table_productivity %>% head()

```

How to interpret the new "Employment Multiplier":
High Sales Multiplier (e.g., 3.0) & Low Employment Multiplier (e.g., 0.5):

* Interpretation: The sector is booming in revenue but not hiring proportionally. This usually means high automation, price increases (inflation driven), or utilizing existing staff more (high productivity).

* Low Sales Multiplier & High Employment Multiplier:

* Interpretation: Warning sign. The sector is hiring aggressively but revenue isn't following. This could indicate declining efficiency or bad planning.
